1. Compute the following expressions for a natural number n chosen by the user:


$$(i) \sum_{\substack{i=0\\i is odd}}^{n-1}(i+3)$$

$$(ii) \prod_{\substack{i=0\\i is even}}^{n-1}(i+3)$$

```python
def odd():
    n = int(input("Donner un entier n :"))
    s = 0
    for i in range(n):
        if i%2 ==1:
            s += i +3
        else:
            s = s
    print(s)
```

```python
odd()
```

```python
def prod():
    m = int(input('Donner un nombre n:'))
    p,i = 1,1
    while i < m :
        if i%2==0:
            p *= i+3
        else : p = p
        i +=1
    print(p)
```

```python
def Fibonacci():
    a0 = int(input('Donner la valeur de a_0:'))
    a1 = int(input('Donner la valeur de a_1:'))
    s0 = (a0+a1)%2
    S = []
    for i in range(15):
        a0 = a1
        a1 = s0
        s0 = (a0 + a1)%2
        S.append(s0)
    print(S)
        
```

```python
def somcar1():
    N = int(input("Donner un nombre N :"))
    S = 0
    for i in range(N+1):
        S += i**2
    print(S)
    
def somcar2():
    N = int(input("Donner un nombre N :"))
    S,i = 0,0
    while i < N:
        S += i**2
        i+=1
    print(S)
```
